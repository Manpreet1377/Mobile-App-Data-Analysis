## Find the number of Free and Paid apps
### Create a new column 'payment_type' and categorize all the applications into 'free' or 'paid'

```python
def is_free(price):
    """
    Checks if the app is paid or free
    Params:
    ------
    price: int or float
    
    Returns:
    ------
    free if price is 0 else paid
    """
    
    assert type(price) in [int,float], 'Enter a number'
    assert price>= 0, 'Enter a non negative number'
    
    return 'free' if price == 0 else 'paid'
```

```python
apple['payment_type'] = apple['price'].map(is_free)
```

## Can we create a recommendation column based on "user_rating" and "rating_count_tot"?

```python
def shortlist(user_rating, rating_count):
    '''
    Shortlist apps
    '''
    
    return apple[(apple['user_rating']>user_rating) & (apple['rating_count_tot'] > rating_count)]['track_name'].values
```

## Paid VS free based on genre... Which genre has more number of free apps than paid apps?

```python
#plotting a percentage barchart and compare
new = apple.groupby(['prime_genre', 'payment_type']).size().unstack()
new_apple = new.copy()
new_apple['Final'] = new_apple['free'] + new_apple["paid"]
new_apple['free'] = (new_apple['free']/new_apple['Final']) * 100
new_apple['paid'] = new_apple['paid']/new_apple['Final'] * 100
new_apple.drop('Final', inplace = True, axis = 1)
new_apple.plot(kind = 'barh', stacked = True, figsize = (12,7))
plt.show()
```

#### Almost 99% of Shopping apps are free. Reason could be that they keep the apps free but charge a minimum fee during any purchase which makes up for the cost of maintaining that app
#### only 25% of educational apps are free. Reason could be that the educational apps charge a minimal fee to  provide the necessary study material

## Rating comparison between free and paid apps

```python
#plotting a percentage barchart to compare
rating = apple.groupby(['user_rating', 'payment_type']).size().unstack()
rating['Final'] = rating['free'] + rating["paid"]
rating['free'] = (rating['free']/rating['Final']) * 100
rating['paid'] = rating['paid']/rating['Final'] * 100
rating.drop('Final', inplace = True, axis = 1)
rating.plot(kind = 'barh', stacked = True, figsize = (12,7))
plt.show()
```

#### The user rating overall is not dependant onn the payment type of the app | There are more number of free apps with 0 rating. We could conclude that there are apps which have been built for free but hardly used or promoted
